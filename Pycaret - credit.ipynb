```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pycaret.classification import *
```

```python
clf = setup(data = data, 
            target = 'LoanStatus', 
            normalize = True, 
            low_variance_threshold=0.1, 
            remove_multicollinearity= True, 
            multicollinearity_threshold= 0.8,
            fix_imbalance = True)
```
